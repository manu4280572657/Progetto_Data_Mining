```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_path = 'anabolic_steroids.csv'
data = pd.read_csv(file_path)
data = pd.read_csv(file_path, delimiter=',')

print(data)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Relative Molecular Mass (g/mol)'], kde=True, color='blue')
plt.title('Distribuzione della Massa Molecolare Relativa')
plt.xlabel('Massa Molecolare Relativa (g/mol)')
plt.ylabel('Frequenza')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Usare una palette di colori interessante per le barre
sns.barplot(data=data, x='Medicinal Use', y='Relative Molecular Mass (g/mol)', estimator='mean', palette='Blues')

# Ruotare le etichette lungo l'asse x per una migliore leggibilità
plt.xticks(rotation=90, ha='right', fontsize=12)

# Titolo e etichette con dimensioni e stile migliorato
plt.title('Massa Molecolare Media per Uso Medicinale', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Uso Medicinale', fontsize=14, labelpad=10)
plt.ylabel('Massa Molecolare Media (g/mol)', fontsize=14, labelpad=10)

# Aggiungere bordi neri alle barre
for bar in plt.gca().patches:
    bar.set_edgecolor('black')
    bar.set_linewidth(1.5)

# Migliorare la visibilità e la disposizione delle etichette
plt.tick_params(axis='both', which='major', labelsize=12)

# Ottimizzare il layout per evitare sovrapposizioni
plt.tight_layout()

# Mostrare il grafico
plt.show()
```

```python
# Ordina i dati per Massa Molecolare in modo decrescente e seleziona i primi 10
top_10_molecular_mass = data.sort_values(by='Relative Molecular Mass (g/mol)', ascending=False).head(10)

# Grafico a barre per i primi 10 composti
plt.figure(figsize=(14, 7))

# Usare una palette di colori più interessante
sns.barplot(x='Original Name', y='Relative Molecular Mass (g/mol)', data=top_10_molecular_mass, palette='viridis')

# Ruotare le etichette per una migliore leggibilità
plt.xticks(rotation=45, ha='right', fontsize=12)

# Titolo e etichette più stilizzate
plt.title('Top 10 Compositi Anabolizzanti per Massa Molecolare', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Nome del Composto', fontsize=14, labelpad=10)
plt.ylabel('Massa Molecolare (g/mol)', fontsize=14, labelpad=10)

# Migliorare la visibilità delle etichette
plt.tick_params(axis='both', which='major', labelsize=12)

# Assicurarsi che tutte le etichette siano visibili e non sovrapposte
plt.tight_layout()

# Aggiungere un bordo alle barre per farle risaltare
for bar in plt.gca().patches:
    bar.set_edgecolor('black')
    bar.set_linewidth(1.5)

# Mostra il grafico
plt.show()
```

```python
# Creiamo una nuova colonna "Anno di Introduzione" estraendo l'anno dalla colonna 'History'
# Se la colonna History contiene informazioni, estrai l'anno (prima si cerca la data, poi si converte)
data['Anno di Introduzione'] = data['History'].str.extract(r'(\d{4})')

# Convertiamo in formato numerico
data['Anno di Introduzione'] = pd.to_numeric(data['Anno di Introduzione'], errors='coerce')

# Raggruppiamo per anno e contiamo il numero di composti per ogni anno
yearly_usage_count = data.groupby('Anno di Introduzione').size()

# Creiamo il grafico a linee per visualizzare la distribuzione nel tempo
plt.figure(figsize=(12, 8))
sns.lineplot(x=yearly_usage_count.index, y=yearly_usage_count.values, marker='o', color='b')

# Personalizzazioni del grafico
plt.title('Distribuzione dell\'Uso Medicinale nel Tempo', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Anno di Introduzione', fontsize=14, labelpad=10)
plt.ylabel('Numero di Composti', fontsize=14, labelpad=10)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

# Mostriamo il grafico
plt.tight_layout()
plt.show()
```
