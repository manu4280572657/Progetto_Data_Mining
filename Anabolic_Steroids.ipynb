```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_path = 'anabolic_steroids.csv'
data = pd.read_csv(file_path)
data = pd.read_csv(file_path, delimiter=',')

print(data)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Relative Molecular Mass (g/mol)'], kde=True, color='blue')
plt.title('Distribuzione della Massa Molecolare Relativa')
plt.xlabel('Massa Molecolare Relativa (g/mol)')
plt.ylabel('Frequenza')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='Medicinal Use', y='Relative Molecular Mass (g/mol)', estimator='mean')
plt.xticks(rotation=90)
plt.title('Massa Molecolare Media per Uso Medicinale')
plt.xlabel('Uso Medicinale')
plt.ylabel('Massa Molecolare Media (g/mol)')
plt.show()
```
